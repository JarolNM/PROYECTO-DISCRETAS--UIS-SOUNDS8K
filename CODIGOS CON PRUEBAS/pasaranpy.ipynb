```python
import os
import numpy as np
import librosa
from tqdm import tqdm

def colapsar_folds_a_bloques(ruta_base):
    folds = [d for d in os.listdir(ruta_base) if os.path.isdir(os.path.join(ruta_base, d)) and 'fold' in d]
    
    for fold in folds:
        ruta_fold = os.path.join(ruta_base, fold)
        archivos_wav = [f for f in os.listdir(ruta_fold) if f.endswith('.wav')]
        
        lista_audios = []
        lista_labels = []
        
        print(f"Empaquetando {fold}...")
        for archivo in tqdm(archivos_wav):
            # Cargar audio (ya está listo según me dices)
            audio, _ = librosa.load(os.path.join(ruta_fold, archivo), sr=None)
            lista_audios.append(audio)
            
            # Extraer etiqueta del nombre (ejemplo: 7383-3-0-1 -> el '3')
            # Ajusta el índice según cómo esté nombrado tu archivo
            label = int(archivo.split('-')[1]) 
            lista_labels.append(label)
            
        # Convertir a arrays de numpy
        datos_X = np.array(lista_audios, dtype=np.float32)
        datos_Y = np.array(lista_labels, dtype=np.int32)
        
        # Guardar un solo archivo de datos y uno de etiquetas por fold
        np.save(os.path.join(ruta_base, f"{fold}_X.npy"), datos_X)
        np.save(os.path.join(ruta_base, f"{fold}_Y.npy"), datos_Y)
        print(f"Guardado: {fold}_X.npy con forma {datos_X.shape}")

# Ejecutar
RUTA_PROYECTO = r"F:\UIS\2026-1\TRATAMIENTO DE SENALES DISCRETAS\PROYECTO\DATOS DE PRUEBA"
colapsar_folds_a_bloques(RUTA_PROYECTO)
```
